# Import Libraries and Data

```python
#standard libraries
import numpy as np
import pandas as pd
#data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
#statistics librarie
import scipy.stats as stats
```

```python
data = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRxp-hsw61ck9cpt5ExZIUhvSGT5zYzoUJl_7xr7YF5qZeazC8393q2taqwkDvMxVirgFDJm1WX3uAC/pub?output=csv'
df = pd.read_csv(data)
df.head()
```

```python
#create a varirable with only women
females = df.loc[df['Gender']=='Female'].copy()
#create variable with only womens height
data = females['Height']
data
```

```python
#creating stats variables for easy calculations
mean = np.mean(data)
std = np.std(data)
min = np.min(data)
max = np.max(data)

x_range = np.linspace(min, max, 1000)
```

#   What is the probability that a female's height is between 55.0 and 56.0 inches

```python
#setting high and low end for calculation
low_end = 55.0
high_end = 56.0 
#calculating pdf 
pdf = stats.norm.pdf(x_range, loc=mean, scale=std)
```

```python
#plotting data with pdf
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(x=data, stat='density')
ax.plot(x_range, pdf)
ax.set_title('PDF of Female Height')
```

```python
#calculate cdf
cdf = stats.norm.cdf(x_range, loc=mean, scale=std )
```

```python
#plotting cdf
fig, ax = plt.subplots(figsize=(12,6))
sns.histplot(x=data, stat='density', cumulative = True)
ax.plot(x_range,cdf)
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
ax.set_title('Range of Interest');
```

```python
#calculating probilility for highlighted range
solution = stats.norm.cdf(high_end, loc=mean,scale=std) - stats.norm.cdf(low_end,loc=mean, scale=std)
print(f'The probilitly of a females height being between 55-56 inches is {solution}')

```

#   What is the probability that a female's height is less than 5 feet (60 in)

```python
#calculating cdf for new question
cdf = stats.norm.cdf(60, loc=mean, scale=std)
print(f'The Probability of a Females Height being less than 5 feet is {cdf}')
```
